```python
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/dataset/data_penjualan_mobil.csv", names=["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"])
df.head(10)
```

```python
#FILTERING FROM ENGINE-SIZE
#df[df['engine-size']>200]
#df[(df['engine-size']>100) & (df['engine-size']>300)]
len(df[df['engine-size'] > 200])
```

```python
#UPDATE TIPE DATA DAN UBAH MISSING VALUE

df['price'] = df['price'].replace(to_replace ="?",value =0)
df['price'] = df['price'].astype("int64")
```

```python
#FILTERING FROM PRICE
df[(df['price']>10000) & (df['price']<20000)]
```

```python
#NORMALISASI

#SIMPLE FEATURE SCALLING
(df['length']) / (df['length'].max())

#min max = xi - x.min / x.max - x.min
(df['length'] - df['length'].min()) /  (df['length'].max() - df['length'].min())

#z score = xi - mean / std
(df['length'] - df['length'].mean()) / df['length'].std()
```

```python
#VISUALISASI DATA

import matplotlib.pyplot as plt

plt.scatter(df['engine-size'], df['price'])
plt.xlabel("Engine-size")
plt.ylabel("Price")
plt.show()
```

```python
#KORELASI

import scipy

x, y = scipy.stats.pearsonr(df['engine-size'], df['price'])
print("Person Coef :" + str(x))
```
